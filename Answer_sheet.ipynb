then $det(A-λI) = 0 ⟺ (λ - a)(λ -d) - bc = 0 ⟺ λ^2 - (a+d)λ + (ad - bc) = 0$ $(1)$ 

To solve $λ$, we first have to obtain $Δ$ from $(1)$, which is $Δ = (a+d)^2 - 4(ad-bc) = (a-d)^2 + 4bc$

1.   $\lambda_1, \lambda_2\in\mathbb R, \lambda_1\ne \lambda_2(Δ > 0)$:

    $$
      \textbf{x}(t) = c_1 v_1 e^{\lambda_1 t} + c_2 v_2 e^{\lambda_2 t} 
    $$

    $$
      \text{($v_1$, $v_2$ are eigenvectors associated with the eigenvalues $\lambda_1$ and $\lambda_2$).}
    $$

  * $\lambda_0$ has 2 linearly independent eigenvectors $v_1$ and $v_2$:

    $$
      \textbf{x}(t) = (c_1 v_1 + c_2 v_2) e^{\lambda_0 t} 
    $$

    $$
      \textbf{x}(t) = [ c_1 v_1 + c_2 (t v_1 + v_2) ] e^{\lambda_0 t} 
    $$

3. $\lambda_1, \lambda_2\in\mathbb C, \lambda_1= \frac{a+d}{2} + \frac{i\sqrt[]{\lvert Δ \rvert}}{2}, \lambda_2= \frac{a+d}{2} - \frac{i\sqrt[]{\lvert Δ \rvert}}{2}(Δ < 0)$:

      \text{($v_1$, $v_2$ are real and imaginary parts of eigenvectors associated with the eigenvalues $\lambda_1$ or $\lambda_2$; $α = \frac{a+d}{2}, β = \frac{i\sqrt[]{\lvert Δ \rvert}}{2}$).}

```python
#Explicit Euler Algorithm
def explicit_euler(f: Function, t: float, x: Vector, h: float) -> Vector:
  """
  The approximate of the exact solution to the ODE dx/dt = f(t, x) at
  time t + h, x is the value of the solution at time t, h is the time step

  t: real number
  h: real number
  x: 2-dimensional real vector
  f: 2-dimensional real-vector-valued function 
  """
  return x + f(t,x)*h
```

```python
#The following functions are used to write the implicit Euler algorithm
def jacobian_x(f: Function, t: float, x: Vector, eps=1E-5) -> Matrix:
  """
  Return the numerical approximate of the Jacobian matrix of the
  2-dimensional real-vector-valued function f(t, x) with respect to x

  t: real number
  x: 2-dimensional real vector
  eps: infinitesimal constant of the approximation
  """
  n = len(x)
  jac = np.zeros((n, n))
  for j in range(n):
    x_eps = np.array(x)
    x_eps[j] += eps
    jac[:, j] = (f(t, x_eps) - f(t, x)) / eps
  return jac

def solve_system(A: Matrix, b: Vector) -> Vector:
  """
  Solution x to the system of linear equations Ax = b

  x: 2-dimensional real vector
  A: 2x2 matrix with real entries
  b: 2-dimensional real vector

  If the system does not have a solution, x will be the vector 0.
  """
  try:
    something = np.linalg.pinv(A)
    return np.dot(something,b)
  except np.linalg.LinAlgError:
    return np.zeros_like(b)

#Implicit Euler Algorithm
def implicit_euler(f: Function, t: float, x: Vector, h: float) -> Vector:
  """
  The approximate of the exact solution to the ODE dx/dt = f(t, x) at
  time t + h, x is the value of the solution at time t, h is the time step

  t: real number
  h: real number
  x: 2-dimensional real vector
  f: 2-dimensional real-vector-valued function 
  """
  y = np.array(x)
  for _ in range(30): #You can change this number of Newton steps!
    A = np.eye(len(y)) - h * jacobian_x(f, t, y)
    b = y - x - h * f(t, y)
    y -= solve_system(A, b)
  return y
```
